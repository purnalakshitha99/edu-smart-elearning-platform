```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the saved model and tokenizer
model = GPT2LMHeadModel.from_pretrained("./gpt2-fine-tuned-mcq")
tokenizer = GPT2Tokenizer.from_pretrained("./gpt2-fine-tuned-mcq")

# Set pad_token_id
model.config.pad_token_id = model.config.eos_token_id

def generate_mcq(context):
    input_text = f"""
    Given the following context, generate a multiple-choice question (MCQ) with four options and the correct answer.

    Context: {context}
    """

    # Tokenize input with attention mask
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]

    # Generate output
    output_ids = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_length=150,
        num_beams=5,
        early_stopping=True
    )

    # Decode the generated text
    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    print("Generated Text:", generated_text)  # Debugging output

    # Extract the question, options, and answer
    try:
        question_start = generated_text.find("Question:")
        options_start = generated_text.find("Options:")
        answer_start = generated_text.find("Answer:")

        if question_start == -1 or options_start == -1 or answer_start == -1:
            return {"error": "The expected labels (Question, Options, Answer) were not found in the output."}

        question = generated_text[question_start + len("Question:"):options_start].strip()
        options_text = generated_text[options_start + len("Options:"):answer_start].strip()
        options = [opt.strip() for opt in options_text.split(",")]
        answer = generated_text[answer_start + len("Answer:"):].strip()

        if len(options) != 4:
            return {"error": "The generated options count is not 4."}

        return {"question": question, "options": options, "answer": answer}

    except Exception as e:
        return {"error": f"Failed to parse the generated text. Error: {str(e)}"}

# Example context
context = "Cloud computing allows users to access and store data over the internet. It provides on-demand availability of computing resources such as servers, storage, and databases."

mcq = generate_mcq(context)
print("Generated Question:", mcq.get("question", "Error"))
print("Options:", mcq.get("options", "Error"))
print("Answer:", mcq.get("answer", "Error"))
```
